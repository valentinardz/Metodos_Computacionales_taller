# Funcion para k^2

\begin{equation}
k^2=\frac{4ar\sin\theta}{a^2+r^2+2ar\sin\theta}
\end{equation}




````python
#k es una funcion de r,a y theta

def k_square(r, theta, a):  
  num = 4*a*r*np.sin(theta)
  den = a**2+r**2+2*a*r*np.sin(theta) 
  return num/den
````

Evaluando la funcion K^2 para un r y una a partucular, tenemos k^2 como función que depende solo de theta.

````python
theta = np.linspace(0, np.pi/2, 100) #valores de theta

plt.figure(figsize=(10,7))
plt.plot(theta, k_square(1.1,theta, 1 ))#evaluamos k para r=1.1 y a=1
plt.xlabel("$\\theta$")
plt.title('$K^2 vs \\theta$')
plt.ylabel("$k^2(\\theta)$")
````

\begin{equation}
K(k)=\int_{0}^{\pi/2} \frac{\mathrm{d\phi}}{\sqrt{1-k^2\sin^2\phi}}
\end{equation}



\begin{equation}
K1(k)=\frac{\mathrm{1}}{\sqrt{1-k^2\sin^2\phi}}
\end{equation}


Para cada valor que tome K^2 se obtiene una función K1 diferente que depende solo de phi, por tanto también una valor diferente de la integral (K).

````python
K1 = lambda k, phi: 1./np.sqrt( 1 - (k * np.sin(phi)) **2 ) #estamos eelvando k^2
````

````python



phi = np.linspace(0,np.pi/2, 100  ) #intervalo de integracion

plt.figure(figsize=(10,7))
plt.title('$K1(\\phi)$')

#se evalua la funcion K1 para el valor que toma la funcion k^2 en (1.1,pi/3,1)============

theta=np.pi/3

k2 = k_square(1.1,theta , 1)# valor de k**2 en el punto (1.1,pi/3,1)
plt.plot(phi, K1( np.sqrt(k2) , phi ), label=f"$\\theta={'π/3'}$")#grafica de la funcion




#se evalua la funcion K1 para el valor que toma la funcion k^2 en (1.1,pi/4,1)============

theta = np.pi/4
k2 = k_square(1.1,theta , 1)# evaluar la funcin k^2 en un punto (1.1,pi/4,1)
plt.plot(phi, K1( np.sqrt(k2) , phi ), label=f"$\\theta={'π/4'}$")
plt.legend()
````

\begin{equation}
K(k)=\int_{0}^{\pi/2} \frac{\mathrm{d\phi}}{\sqrt{1-k^2\sin^2\phi}}
\end{equation}


Bucamos el valor de K, es decir la integral de k1 entre 0 a π/2 al tomar un valor de k^2

````python

#Solucion de la Integral empleando el metodo de trapecio=====================================

# para un valor theta, r, a 

phi = np.linspace(0, np.pi/2, 100 )#intervalo de integracion 

theta = np.pi/3
r = 1.1
a = 1
k2 = k_square(r, theta , a)# valor de k**2 en el punto 1.1,pi/2,1
y = K1( np.sqrt(k2) , phi ) #k1 como funcion de phi
integrate.trapz(y, phi )
````

Para cada valor que tome K^2 se obtiene una función E1 diferente que depende solo de phi, por tanto también una valor diferente de la integral (E).


\begin{equation}
E(k)=\int_{0}^{\pi/2}  \mathrm{d\phi} \sqrt{1-k^2\sin^2\phi}

\begin{equation}
E1(k)= \sqrt{1-k^2\sin^2\phi}

````python
E1 = lambda k, phi: np.sqrt( 1 - (k * np.sin(phi)) **2 ) # E1  estamos elevando k^2
````

````python
#Solucion de la Integral empleando el metodo de trapecio=====================================


phi = np.linspace(0,np.pi/2, 100  )#intervalo de integracion
plt.figure(figsize=(10,6))
plt.title('$E1(\\phi)$') 

#se evalua la funcion E1 para el valor que toma la funcion k^2 en (1.1,pi/3,1)============

theta = np.pi/3
k2 = k_square(1.1,theta , 1)                      
plt.plot(phi, E1( np.sqrt(k2) , phi ), label=f"$\\theta={'π/3'}$")

#se evalua la funcion K1 para el valor que toma la funcion k^2 en (1.1,pi/4,1)============

theta = np.pi/4
k2 = k_square(1.1,theta , 1)       # valor k**2 en un punto (1.1,pi/4,1)
plt.plot(phi, E1( np.sqrt(k2) , phi ), label=f"$\\theta={'π/4'}$")# E1 dependiendo solo de phi
plt.legend()
````

\begin{equation}
E(k)=\int_{0}^{\pi/2}  \mathrm{d\phi} \sqrt{1-k^2\sin^2\phi}  
\end{equation}


Bucamos el valor de E, es decir la integral de E1 entre 0 a π/2 al tomar un valor de k^2

````python

#Solucion de la Integral empleando el metodo de trapecio==============================

# para un valor theta, r, a 

phi = np.linspace(0, np.pi/2, 100 )    #intervalo de integracion
theta = np.pi/3
r = 1.1
a = 1
k2 = k_square(r, theta , a)             #valor de k^2 en el punto 1.1,pi/3,1
y = E1( np.sqrt(k2) , phi )             # E1 como funcion phi
integrate.trapz(y, phi )                # valor de la intergal de E1 
````


\begin{equation}
A_{\phi}(r, \theta)=\frac{\mu_0}{4\pi}\frac{4Ia}{\sqrt{a^2+r^2+2ar\sin \theta}}\left(\frac{(2-k^2)K(k)-2E(k)}{k^2} \right)
\end{equation}


Al evaluar r,theta,a en  de K^2 se obtien un valor (cte)
al reemplazar este valor en E1 y en K1, entonces estas funciones dependen solo de phi
Al evaluar la integral de E1 y K1 en el intervalo asigando, entonces K y E toman un valor cte.


\begin{equation}
\left(\frac{(2-k^2)K(k)-2E(k)}{k^2} \right)
\end{equation}

Este termino es cte en la funcion A=====> A es funcion de a,r,theta

````python

#cte=1 es u0 /4pi

def Apotencial(r, theta, phi, a, I, cte = 1 ):
  
  T11 = cte*4*I*a
  T12 = np.sqrt( a**2 + r**2 + 2*a*r*np.sin(theta))
  T1 = T11/T12

  k2= k_square(r, theta, a) #valor de k^2 para r,theta,a
 
  T21 = (2 - k2)
  T22 = integrate.trapz(  K1(np.sqrt(k2), phi), phi)#funcion K integral de K1
  T2 = T21*T22

  T3 =  integrate.trapz(  E1(np.sqrt(k2), phi), phi) #funcion E integral de E1
  A = T1*(T2 - 2*T3)/k2

  return A

  
````

````python
# Parametros===========================================================================

theta = np.pi/3
r = 1.1
a = 1.0
I = 3.0
phi = np.linspace(0, np.pi/2, 100)#intervalo de integracion para K1 y E1

#valor del potencial en el punto  (r, θ, φ) cuando la espira tiene radio a y corriente I

A = Apotencial(r, theta, phi, a, I, cte = 1 )
print(A)
````

````python
r = 1.1                    # Radio del punto donde calcular el potencial 
a = 1.0                    # Radio de la espira, unidades computacionales
I = 3.0                    #Corriente que circula por la espira, unidades computacionales
N = 100                    # Numero o resolución en theta y phi

#Tomando estos valores entoces A es funcion de Theta

theta = np.linspace(0.01, np.pi/2, N)

phi = np.linspace(0, np.pi/2, N)#intervalo de integracion para k1 y E1

Ap = np.zeros(len(theta))#array donde se guardan lo valores del potencial

#para cada posicion i en ap, se guarda el valor de A cuando theta toma en valor t 

for i, t in enumerate(theta): 
    Ap[i] = Apotencial(r, t, phi, a, I, cte = 1 )
  
````

````python
plt.figure(figsize=(12,9))
plt.plot(theta, Ap)
plt.xlim(0.1, 2)
plt.ylim(0,20)
plt.xlabel("$\\theta$")
plt.ylabel("$A(r=1.1, \\theta)$")
````
